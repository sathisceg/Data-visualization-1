```python
from wordcloud import WordCloud, STOPWORDS 
import matplotlib.pyplot as plt 
import pandas as pd 


df = pd.read_excel("temp_data2.xlsx", "Sheet1")

comment_words = ' '
stopwords = set(STOPWORDS) 


i=0

for index, row in df.iterrows():
    
    
    summary=row['summary']
    motive=row['motive']
    
    if type(summary)==float:
        continue

    if type(motive)==float:
        continue
        
        
    val = str(summary[10:])+str(motive)

    if(i==10):
        break
    
    i=i+1
    tokens = val.split()
       
    for i in range(len(tokens)): 
        tokens[i] = tokens[i].lower() 
        
    for words in tokens: 
        comment_words = comment_words+words+" "   


```

```python
wordcloud = WordCloud(width = 800, height = 800,background_color ='white',stopwords = stopwords,min_font_size = 10).generate(comment_words) 



plt.figure(figsize = (8, 8), facecolor = None) 
plt.imshow(wordcloud) 
plt.axis("off") 
plt.tight_layout(pad = 0) 

plt.show() 
```

```python
# Reads 'Youtube04-Eminem.csv' file 
df = pd.read_excel("global_terrorism_database.xlsx", "Data") 

list_of_edge=[]

# iterate through the csv file 
for val in df['related']:     
    if type(val)==float:
        continue
        
    temp=val.split(',')
    
#     print(val,type(val),temp[0],len(temp))
    

    
    if(len(temp)>2):
        j=1 
        while(j<len(temp)):
            list_of_edge.append((temp[0],temp[j]))
            j=j+1
    else:
        list_of_edge.append((temp[0],temp[1]))

  
```

```python
list_of_edge
```

```python
import networkx as nx
```

```python
g=nx.Graph()
```

```python
g.add_edges_from(list_of_edge)
```

```python
nx.draw(g)
```
