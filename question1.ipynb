```python
date_incidents = pd.read_csv("baltimore_crimes.csv")

crime_date_data={}

for index, row in date_incidents.iterrows():
#     print(index,row['CrimeDate'],row['Total Incidents'])
    tdate=str(row['CrimeDate']).split('/')
    
    dayofweek = datetime.date(int(tdate[2]),int(tdate[0]),int(tdate[1])).strftime("%A")
#     print(dayofweek)
    if dayofweek in crime_date_data:
        crime_date_data[dayofweek]+=row['Total Incidents']
    else:
        crime_date_data[dayofweek]=row['Total Incidents']
    
    
    
day_list=list(crime_date_data.keys())


count_list=list(crime_date_data.values())

```

```python
import pandas as pd


import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
district_incidents = pd.read_csv("baltimore_crimes.csv", usecols=['CrimeCode', 'Location','Total Incidents'],)



crime_data={}



for index, row in district_incidents.iterrows():
#     print(row['CrimeCode'],row['Location'],row['Total Incidents'])


    
    temp=str(row['CrimeCode'])+"#"+str(row['Location'])



    if temp in crime_data:
    	crime_data[temp]+=row['Total Incidents']
    else:
    	crime_data[temp]=row['Total Incidents']

   

```

```python
crimecode_list=[]
location_list=[]
count_list=[]


for k,v in crime_data.items():
    
    c=k.split("#")[0]
    l=k.split("#")[1]    
    crimecode_list.append(c);
    location_list.append(l);
    count_list.append(v);
```

```python
data = {'code': crimecode_list,
         'loc': location_list,
         'count': count_list
         }
df = pd.DataFrame.from_dict(data)
```

```python
input_data=df.pivot('code','loc','count')


plt.subplots(figsize=(20,15))
sns.heatmap(input_data)
```
